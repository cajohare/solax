```python
import sys
sys.path.append('../src')
from Params import *
from PlotFuncs import *
from Like import *
from AxionFuncs import *  
```

```python
fig,ax = MySquarePlot(r'$\omega$ [keV]',r'$P_{\gamma \rightarrow a}$')


m_a = 1.0e-1
Bfield = 3.5
Length = 20
T_operating = 1.8
pressure_max = (m_a)**2.0*T_operating/0.02
g = 1e-10



E = logspace(-6,log10(20.0),1000)
n_ex = 6
po = logspace(-7,0,n_ex)

for i in range(0,n_ex-1):
    pressure = pressure_max*po[i]

    
    # P_AX calculation:
    m_gam = sqrt(0.02*pressure/T_operating) # eV
    gam = 0.29*pressure/(E**3.1*T_operating) # m^-1
    gam_eV = gam*1.97e-7 # eV
    L_eV = Length/1.97e-7 # eV^-1
    B = Bfield*(1e-19*195)
    normq = L_eV/(4*1000)
    q = (m_a**2.0-m_gam**2.0)/(2*E*1000) # eV
    P_ax = 0.25*(B**2.0)*1.0/(q**2.0+gam_eV**2.0/4)*(1+exp(-gam*Length)-2*exp(-gam*Length/2.0)*cos(q*L_eV))

    # Plot it
    plt.loglog(E,P_ax,'-',label=po[i])

    # energy at which oscillations start:
    a = (0.29*pressure/T_operating)*Length
    wosc = (2*log(2)/a)**(-1/3.1)
    plt.axvline(wosc)

    # Cross over in low/high energy scaling regimes:
    wcross = (1.97e-7*(0.29*pressure/T_operating)*1000/((m_a)**2.0-(m_gam)**2.0))**(1/2.1)
    plt.axvline(wcross,color='k')

    
plt.legend(title=r'pressure/max pressure')

plt.plot(E,1.0e-13*E**6.2,'k-',lw=3) # high energy scaling
plt.plot(E,1.0e-17*E**2,'k-',lw=3) # low energy scaling

```
